```python
import pandas
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
heart = pandas.read_csv("heart.csv")
```

```python
print(heart.columns)
print(heart.shape)
```

### Visualize all columns

```python
heart.hist(figsize= (20,20))
plt.show
```

```python
heart.index
```

### Showing a heat map to know each column effect on the target

```python
corrmat = heart.corr()
fig = plt.figure(figsize = (12,9))

sns.heatmap(corrmat, vmax=1 , square = True)
plt.show()
```

### Preparing the features and the target and split the train and the test data

```python
features = heart.columns[heart.columns != 'target']
target = 'target'

from sklearn.model_selection import train_test_split

train = heart.sample(frac=0.8, random_state=1)

test = heart.loc[~heart.index.isin(train.index)]
```

### Trying the first model which is the random forest regressor from sklearn

```python
from sklearn.metrics import mean_squared_error, classification_report, accuracy_score
from sklearn.ensemble import RandomForestRegressor

rfr = RandomForestRegressor(n_estimators = 100, min_samples_leaf=10, random_state = 1)

rfr.fit(train[features],train[target])

pred = rfr.predict(test[features])
pred[pred>0.5] = 1
pred[pred<0.5] = 0
print(mean_squared_error(pred,test[target]))
print(accuracy_score(test[target], pred))
print(classification_report(test[target], pred))
```

### Applying the linear regression to the data

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(train[features],train[target])

pred = lr.predict(test[features])
pred[pred>0.5] = 1
pred[pred<0.5] = 0
print(mean_squared_error(pred,test[target]))
print(accuracy_score(test[target], pred))
print(classification_report(test[target], pred))
```

### Applying the SVM to the data

```python
from sklearn import svm, datasets

C = 1.0
svc = svm.SVC(kernel='linear', C=C).fit(train[features],train[target])
pred = svc.predict(test[features])
pred[pred>0.5] = 1
pred[pred<0.5] = 0
print(mean_squared_error(pred,test[target]))
print(accuracy_score(test[target], pred))
print(classification_report(test[target], pred))
```
